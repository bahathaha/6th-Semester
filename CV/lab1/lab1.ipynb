```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = "/home/nitdelhi/Desktop/231210033/6th-Semester/CV/lab1/IRIS.csv"
df = pd.read_csv(path)
```

```python
df.head()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,10))
df.boxplot(by='species', column=['sepal_length'], grid=False, ax=axes[0,0])
df.boxplot(by='species', column=['sepal_width'], grid=False, ax=axes[0,1])
df.boxplot(by='species', column=['petal_length'], grid=False, ax=axes[1,0])
df.boxplot(by='species', column=['petal_width'], grid=False, ax=axes[1,1])
plt.show()
```
